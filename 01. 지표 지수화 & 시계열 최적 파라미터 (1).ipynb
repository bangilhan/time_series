# 지수 시계열 예측

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
#from pmdarima.arima import auto_arima ## ADP 볼 때는 없을 패키지
import itertools # 내장 패키지
import matplotlib.pyplot as plt
%matplotlib inline

p = range(0,3)
d = range(0,2)
q = range(0,3)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12 # 계절성이 7일마다 있다고 생각해서 7 입력. 얘도위에 range로 탐색해도 됨
 ) for x in list(itertools.product(p, d, q))]
```

```python
def future_x(x):


    #train_data, test_data =train_test_split(df_x['population'], test_size=0.2, shuffle=False)
    
    aic = []
    params = []
    for i in pdq:
        for j in seasonal_pdq:
            try:
                 model = SARIMAX(x.values, order=(i), seasonal_order = (j
                ))
                 model_fit = model.fit()
                 print(f'SARIMA: {i}{j} >> AIC : {round(model_fit.aic,2)}')
                 aic.append(round(model_fit.aic,2))
                 params.append((i, j))
            except:
                 continue

    # Search optimal parameters
    optimal = [(params[i], j) for i, j in enumerate(aic) if j == min(aic)]
    
    print('optimal_{0}'.format(x))
    print('{0}'.format(optimal))
    # small pdq는 (1,1,0), large pdq는 (1,1,2) 그리고 Seasonal parameter는 12인 것을볼 수 있다.


    model_opt = SARIMAX(x, order=optimal[0][0][0], seasonal_order= (optimal[0][0][1]))
    model_opt_fit = model_opt.fit()
    model_opt_fit.summary()
    # ARIMA보다 SARIMA가 AIC가 훨씬 낮은 것을 볼 수 있다.


    ahead=pd.date_range(start='2021-01-01', end='2023-12-01', freq='MS')
    
    prediction = model_opt_fit.get_forecast(36)

    predicted_value = prediction.predicted_mean

    predicted_lb=prediction.conf_int()
    predicted_ub=prediction.conf_int()
    predict_index= ahead
    
    return predicted_value,optimal, print(x)
```
